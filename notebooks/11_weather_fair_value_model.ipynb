# Weather Fair-Value Model

Train and inspect the first logistic fair-value model once enough resolved rows exist.

## Finance Feature Roadmap For Fair Value

Use `forecast_event_indicator` as the baseline weather-implied event feature, then train calibrated probabilities with finance-specific inputs:

- `market_mid`, `market_microprice`, `market_spread`, `market_depth_imbalance`
- `forecast_minus_threshold`, `abs_forecast_distance_to_threshold`, `forecast_event_indicator`
- `time_to_expiry_hours`, forecast age, contract type, comparison operator
- partition-level features: sum of YES bids/asks, basket richness, neighboring range prices

The model target remains contract settlement label, but the strategy should consume executable edges versus bid/ask, not midpoint edge.
